```python
# Loading the nursery.data file, assigning the provided column names,
# showing a preview, and saving to CSV for you to download.
import pandas as pd
#from caas_jupyter_tools import display_dataframe_to_user

path = 'nursery.data'
cols = ['parents','has_nurs','form','children','housing','finance','social','health','class']

# Try reading with comma separator first, fallback to whitespace if needed
try:
    df = pd.read_csv(path, header=None, names=cols, sep=',', engine='python')
except Exception as e:
    # fallback: read lines and split manually
    with open(path, 'r', encoding='utf-8') as f:
        lines = [ln.strip() for ln in f if ln.strip()]
    parts = [ln.split(',') for ln in lines]
    maxc = max(len(p) for p in parts)
    if maxc == 9:
        df = pd.DataFrame(parts, columns=cols)
    else:
        # make best-effort: truncate or pad to 9 columns
        cleaned = [p[:9] + ['']*(9 - len(p)) for p in parts]
        df = pd.DataFrame(cleaned, columns=cols)

# Quick sanity checks and conversions
rows, cols_count = df.shape
print(f'Loaded file: {path}')
print(f'Rows: {rows}, Columns: {cols_count}')
print('\nColumn names:')
print(list(df.columns))

# Show first 200 rows in an interactive table for quick inspection
#display_dataframe_to_user('Nursery dataset (first 200 rows)', df.head(200))

# Save to CSV
out_path = 'nursery.csv'
df.to_csv(out_path, index=False)
print(f'\nSaved CSV to: {out_path}')
```
